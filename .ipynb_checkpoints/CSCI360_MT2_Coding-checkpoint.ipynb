#### Step 1: Load the digits dataset. see details here (https://scikit-learn.org/1.5/modules/generated/sklearn.datasets.load_digits.html)

```python
### Your Code here
from sklearn.datasets import load_digits
digits = load_digits(as_frame=True)
print(digits.data.shape)
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(digits.images[455])
plt.show()
print(digits.target[455])
```

#### Step 2: Split the dataset into training and testing sets

```python
### Your code here
X_Train = digits.data.iloc[0:1258]
Y_Train = digits.target.iloc[0:1258]
X_Test = digits.data.iloc[1258:1797]
Y_Test = digits.target.iloc[1258:1797]
```

#### Step 3: Use MLPClassifier to build the neural network and train

```python
### Your code here
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', solver='adam').fit(X_Train, Y_Train)
```

#### Step 4: Evaluate the model and report test accuracy

```python
### Your code here
from sklearn.metrics import classification_report
print("Classification Report for Test Set:\n", classification_report(Y_Test, clf.predict(X_Test)))
print("Classification Report for Training Set:\n", classification_report(Y_Train, clf.predict(X_Train)))
```

#### Step 5: Visualize. Display a few test images alongside their true and predicted labels

```python
### Your code here
import numpy as np

unique_classes = np.unique(Y_Test)
for cls in unique_classes:
    idx = np.where(Y_Test == cls)[0][0]  
    single_test_instance = X_Test.iloc[[idx]]  
    predicted_label = clf.predict(single_test_instance)[0] 
    plt.matshow(digits.images[1258:1797][idx])
    plt.show()
    print('Predicted Label: ' + str(predicted_label))
    print('True Label: ' + str(cls))

```
